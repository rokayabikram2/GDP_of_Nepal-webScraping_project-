```python
url='https://www.worldometers.info/gdp/nepal-gdp/'
```

```python
import requests
```

```python
html=requests.get(url).content
```

```python
type(html)
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(html,'html.parser')
```

```python
tb=soup.find('tbody')
```

```python
tr=tb.find_all('tr')
```

```python
tr[0]
```

```python

all_data=[]
for index in range(0,len(tr)):
    year_data=[]
    table_row=tr[index]
    table_data=table_row.find_all('td')
    
    for item in table_data:
        # print(item.text)
        year_data.append(item.text)
    
    all_data.append(year_data)
        
```

```python
all_data
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(all_data)
```

```python
df
```

```python
columns_name=['Year',
             'GDP Nominal(Current USD)',
             'GDP Real(Inflation adj.)',
             'GDP change',
             'GDP per capita',
             'pop. change',
             'Population'
            ]
```

```python
df.columns=columns_name
```

```python
df.tail()
```

```python
df.to_csv('GDP_of_Nepal.csv',index=False)
```

```python
df=pd.read_csv('GDP_of_Nepal.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['GDP Nominal(Current USD)'].str.replace('$','',regex=True).replace(",",'',regex=True)
```

```python
df['GDP Nominal(Current USD)'] = df['GDP Nominal(Current USD)'].str.replace('$' , '', regex = True).replace(",", '', regex = True).astype(float)
```

```python
df.dtypes
```

```python
df['GDP Real(Inflation adj.)'] = df['GDP Real(Inflation adj.)'].str.replace('$' , '', regex = True).replace(",", '', regex = True).astype(float)
```

```python
df.dtypes
```

```python
df.info()
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df['Year'],df['GDP Nominal(Current USD)'],color='r',label="GDP USD($)")
plt.legend()
plt.title("GDP of Nepal in US dollar each year")
plt.xlabel("Date")
plt.ylabel("PAisa in dollar")
plt.show()
```

```python
df['GDP change']=df['GDP change'].str.replace('%','',regex=True).astype(float)
```

```python
df.dtypes
```

```python
df['GDP per capita']=df['GDP per capita'].str.replace('$','',regex=True).astype(float)
```

```python
df.dtypes
```

```python
df['pop. change']=df['pop. change'].str.replace("%",'',regex=True).astype(float)
```

```python
df.dtypes
```

```python
df['Population']=df['Population'].str.replace(',','',regex=True).astype(float)
```

```python
df.dtypes
```
